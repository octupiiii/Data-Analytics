# **Dataset Description**
The dataset that we have chosen is a compilation of data collected in 2020 from CDC's Behavioral Risk Factor Surveillance System (BRFSS) in order to identify the main factors for heart disease. As a result, this dataset offers insights into how we can prevent heart complications by identifying certain trends within the data. On a wider scope, the CDC uses the data from the BRFSS to promote healthy living within the United States of America.

# **Description of the Collection Process**
According to the [CDC's 2020 BRFSS Overview](https://www.cdc.gov/brfss/annual_data/2020/pdf/overview-2020-508.pdf) document, adult respondents from all over the United States–including territories like Guam and Puerto Rico–were interviewed and surveyed via landline or cellular telephone. The questionnaire used in the surveys consisted of health-related questions that were either agreed upon by BRFSS administrators and the CDC, or taken from already established American health surveys. Such questions ask about the respondents' conditions, behaviors, or even their demographic information.

To build their sample records, two methods were used. For the landline sample, a disproportionate stratified sample (DSS) design was used. Whereas the cellular telephone sample made use of random sampling.

In this specific survey, DSS was implemented by dividing the telephone numbers into either the high-density or medium-density stratum. These two strata were made based on the likeliness of the number belonging to a household. The former stratum includes multiple groups of 100 numbers, where there is at least one household number present in each group. Whereas the latter stratum includes multiple groups of 100 numbers, where there are no household numbers present in each group.

On the other hand, random sampling was implemented in this survey by gathering numbers from a database, and splitting them into smaller groups. Then, no more than one number randomly picked from each of the groups were called to answer the BRFSS survey.

Although the medium of telephone surveys may raise some questions about data accuracy and reliability, there are a few things that the CDC does to ensure quality data. First of all, all interviewers are formally trained and evaluated for the task of surveying people for the BRFSS–therefore reducing human error. Likewise, Computer-Assisted Telephone Interview (CATI) systems are used as well to streamline the interview process. These systems are also updated yearly so that all features aren't obsolete and usable.

However, data collected in this way is still prone to errors. For example, an interviewee might give an answer that is not representative of their true state, or noise can interfere with the quality of the call. All of these issues can warp the final conclusion, making it more difficult to determine what factors lead to an increased risk of heart disease.

#**Structure of the Dataset File**

The dataset is entitled “Indicators of Heart Disease”. It is derived from the 2020 Behavioral Risk Factor Surveillance System (BRFSS) dataset by the Centers for Disease Control and Prevention (CDC). [The Indicators of Heart Disease](https://www.kaggle.com/datasets/kamilpytlak/personal-key-indicators-of-heart-disease/data) contain variables that may influence heart disease. It is owned by Kamil Pytlak, who performed data processing on the initial data. The dataset is expected to be updated annually.

The original 2020 BRFSS dataset was in SAS format, which was then converted into CSV format by Pytlak. The data is represented using a data matrix with 18 columns and 319,796 rows. The first row of the dataset contains headers, with each column representing a variable with a direct or indirect effect on heart disease. The following rows represent the observations obtained from the data collection. The dataset contains a mix of categorical and numerical data.

Fourteen (14) variables are categorical, 12 of which are nominal and 2 ordinal.  Most of the categorical ordinal variables are characterized by “Yes” or “No” values. The remaining 4 are numerical values, with 3 discrete and 1 continuous.  Overall, the dataset contains a total of 18 variables and 319,795 observations.

## **Inference on Survey Data**

Inference on survey data involves using statistical methods to draw conclusions about a population, where the margin of error shows the accuracy of estimates, the percentage with heart disease reflects the proportion affected, and the minimum and maximum values indicate the range of observed responses.

```python
preprocessed_heart_disease_df['HeartDisease'] = preprocessed_heart_disease_df['HeartDisease'].map({'Yes': 1, 'No': 0})

total_individuals = len(preprocessed_heart_disease_df)
heart_disease_count = preprocessed_heart_disease_df['HeartDisease'].sum()
proportion_heart_disease = heart_disease_count / total_individuals
percentage_heart_disease = proportion_heart_disease * 100
print(f'Percentage with heart disease: {percentage_heart_disease:.2f}%')
```

Percentage with Heart Disease: Approximately 8.56% of the sample has heart disease.


```python
standard_error = np.sqrt(proportion_heart_disease * (1 - proportion_heart_disease) / total_individuals)
z_score = stats.norm.ppf(0.975)
margin_of_error = z_score * standard_error
print('Margin of Error (MoE): {:.4f}%'.format(margin_of_error * 100))
```

Margin of Error: ±0.0970% around the percentage estimate, indicating the precision of the estimate.

```python
maximum_value = (proportion_heart_disease + margin_of_error)
minimum_value = (proportion_heart_disease - margin_of_error)

print('Minimum value: {:.4f}'.format(minimum_value * 100), '\nMaximum value: {:.4f}'.format(maximum_value * 100))
```

Confidence Interval: The true percentage of individuals with heart disease is likely between 8.46% and 8.66%, providing a range of uncertainty around the estimate.

## **Chi-Square**

We use the Chi-Square test because the variables are all categorical. This test helps in finding the significance of the relationship between lifestyle factors and heart disease.

**Hypotheses:**

Null Hypothesis (H₀): There is no significant association between the categorical variable (BMI, Physical Health, Mental Health, Sleep Time) and heart disease. The variables are independent.

Alternative Hypothesis (H₁): There is a significant association between the categorical variable and heart disease. The variables are not independen

**Confidence Level:**

For this test is set at 95%. This means we are using a significance level (𝛼) of 0.05.

```python
contingency_table = pd.crosstab(heart_disease_df['HeartDisease'], heart_disease_df['Smoking'])

print("Smoking Contingency Table:")
print(contingency_table)

chi2, p, dof, expected = chi2_contingency(contingency_table)

print("\nChi-Square Test Results:")
print(f"Chi-Square Statistic: {chi2:.3f}")
print(f"P-Value: {p:.1e}")
print(f"Degrees of Freedom: {dof} \n")

expected_df = pd.DataFrame(expected, index=contingency_table.index, columns=contingency_table.columns)
alpha = 0.05
if p <= alpha:
    print("The variables are not independent (reject the null hypothesis).\n")
else:
    print("The variables are independent (fail to reject the null hypothesis).\n")

print("Expected Frequencies Table:")
print(expected_df.round(2))

if (expected < 5).any():
    print("Warning: Some expected frequencies are less than 5. Results may not be reliable.")
```

The Chi-Square test results for the participants who smokes contingency table show a Chi-Square statistic of 3713.033, indicating a significant difference between observed and expected frequencies. The p-value of 0 confirms a strong association between smoking and heart disease.
